### LINE PLOT

```python

plt.plot(iris["SepalWidthCm"],iris["SepalLengthCm"],color="green",linestyle=":",marker="^",markersize=5)
plt.title("sepal width Vs sepal length of iris species")
plt.xlabel("Sepalwidthcm")
plt.ylabel("sepallengthcm")
plt.grid()
plt.show()
```

### SCATTER PLOT

```python

plt.scatter(x=iris["SepalWidthCm"],y=iris["SepalLengthCm"],color="red",linestyle=":",marker="*",s=30,alpha=0.8)
plt.title("sepal width Vs sepal length of iris species")
plt.xlabel("Sepalwidthcm")
plt.ylabel("sepallengthcm")
plt.grid()
plt.show()
```

### BAR CHART

```python
plt.bar(iris["Species"],iris["PetalLengthCm"])
plt.title("petallength of iris species")
plt.xlabel("Species")
plt.ylabel("Petal Length in cm")
plt.show()
```

```python
plt.bar(iris["Species"],iris["PetalLengthCm"])
plt.title("petallength of iris species")
plt.xlabel("Species")
plt.ylabel("Petal Length in cm")
plt.grid()
plt.show()
```
